确定初始状态和转移矩阵，经过25步模拟后，概率分布趋于稳定状态

```python
# import numpy as np
import matplotlib.pyplot as plt

# 定义状态集合
states = ['A', 'B', 'C', 'D']

# 定义一个新的转移概率矩阵
transition_matrix = np.array([
    [0.5, 0.2, 0.2, 0.1],
    [0.1, 0.6, 0.2, 0.1],
    [0.1, 0.2, 0.5, 0.2],
    [0.4, 0.1, 0.1, 0.4]
])

# 定义一个新的初始状态分布
initial_distribution = np.array([0.25, 0.25, 0.25, 0.25])  # 四个状态的等概率分布

# 模拟马尔可夫链的状态转移
def simulate_markov_chain(transition_matrix, initial_distribution, steps):
    state_distribution = initial_distribution
    chain_history = [state_distribution]
    for _ in range(steps):
        state_distribution = np.dot(state_distribution, transition_matrix)
        chain_history.append(state_distribution)
    return chain_history

# 模拟25步
steps = 25
chain_history = simulate_markov_chain(transition_matrix, initial_distribution, steps)

# 打印每次迭代后的状态分布
for i, state_probabilities in enumerate(chain_history):
    print(f"Step {i}: {state_probabilities}")

# 可视化链的状态分布
plt.figure(figsize=(10, 6))
for i, state in enumerate(states):
    state_probabilities = [step[i] for step in chain_history]
    plt.plot(range(steps + 1), state_probabilities, label=f"State {state}")

plt.xlabel("Step")
plt.ylabel("Probability")
plt.title("State Distribution of Markov Chain Over Time")
plt.legend(ncol=2)  # 使用ncol参数来指定图例的列数
plt.show()
```
