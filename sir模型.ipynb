```python
import numpy as np
import matplotlib.pyplot as plt

def sir_model(infection_rate, recovery_rate, population, infected, days):
    susceptible = population - infected
    S = [susceptible]
    I = [infected]
    R = [0]

    for _ in range(days):
        new_infected = infection_rate * susceptible * infected / population
        new_recovered = recovery_rate * infected
        susceptible -= new_infected
        infected += new_infected - new_recovered
        R.append(R[-1] + new_recovered)
        S.append(susceptible)
        I.append(infected)

    return S, I, R

def plot_sir(S, I, R):
    plt.plot(S, label='Susceptible')
    plt.plot(I, label='Infectious')
    plt.plot(R, label='Recovered')
    plt.xlabel('Days')
    plt.ylabel('Population')
    plt.title('SIR  Model')
    plt.legend()
    plt.show()

# 
infection_rate = 0.3  #  (lambda)
recovery_rate = 0.1  #  (mu)
population = 1000  # 
infected = 1  # 
days = 100

# 
S, I, R = sir_model(infection_rate, recovery_rate, population, infected, days)

# 
plot_sir(S, I, R)
```
