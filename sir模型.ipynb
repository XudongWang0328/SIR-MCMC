```python
import numpy as np
import matplotlib.pyplot as plt

def sir_model(lmbda, mu, population, infected, days):
    susceptible = population - infected
    S = [susceptible]
    I = [infected]
    R = [0]

    for _ in range(days):
        new_infected = lmbda * susceptible * infected / population
        new_recovered = mu * infected
        susceptible -= new_infected
        infected += new_infected - new_recovered
        R.append(R[-1] + new_recovered)
        S.append(susceptible)
        I.append(infected)

    return S, I, R

def plot_sir(S, I, R):
    plt.plot(S, label='Susceptible')
    plt.plot(I, label='Infectious')
    plt.plot(R, label='Recovered')
    plt.xlabel('Days')
    plt.ylabel('Population')
    plt.title('SIR  Model')
    plt.legend()
    plt.show()

# 参数
lmbda = 0.4  # 感染率 (lambda)
mu = 0.2  # 康复率 (mu)
population = 10000  # 总人口
infected = 1  # 初始感染人数
days = 100

# 运行模拟
S, I, R = sir_model(lmbda, mu, population, infected, days)

# 绘制结果
plot_sir(S, I, R)
```
